```python
# Parameters for the sigmoid function
min_fee = 3000  # Minimum fee (0.3%)
max_fee = 10000  # Maximum fee (1.0%)
a = 1.5  # Steepness of the sigmoid
b = 6 # Midpoint (volatility level where curve transitions)

# Define the sigmoid function
def sigmoid(volatility):
    return min_fee + (max_fee - min_fee) / (1 + np.exp(-a * (volatility - b)))
```

```python
# Generate volatility values
volatility_values = np.linspace(0, 10, 100)  # Volatility from 0% to 10%
fee_values = sigmoid(volatility_values)

# Plot the sigmoid curve
plt.figure(figsize=(10, 6))
plt.plot(volatility_values, fee_values, label="Dynamic Fee", color="blue", linewidth=2)
plt.axhline(min_fee, linestyle="--", color="green", label="Min Fee (0.3%)")
plt.axhline(max_fee, linestyle="--", color="red", label="Max Fee (1.0%)")
plt.axvline(b, linestyle="--", color="purple", label="Midpoint (5%)")
plt.title("Sigmoid Function for Dynamic Fee Adjustment", fontsize=14)
plt.xlabel("Volatility (%)", fontsize=12)
plt.ylabel("Fee (Basis Points)", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
